```python
from langgraph.graph import StateGraph, START, END
from typing import TypedDict
```

```python
class batsman_profile(TypedDict):

    run: int
    four: int
    six: int
    ball: int

    sr: float
    bpb: float
    summary: str
```

```python
def calc_sr(state: batsman_profile) -> batsman_profile:

    sr = (state['run']/state['ball'])* 100

    return {'sr': sr}

def calc_bpb(state: batsman_profile) -> batsman_profile:

    bpb = state['ball']/(state['four'] + state['six'])

    return {'bpb': bpb}

def summary(state: batsman_profile) -> batsman_profile:

    summary = f"""
                strike rate: {state['sr']}\n\n\n
                ball per boundary: {state['bpb']}\n\n\n

              """
    
    state['summary'] = summary
    return state
```

```python
graph = StateGraph(batsman_profile)

graph.add_node('calc_sr', calc_sr)
graph.add_node('calc_bpb', calc_bpb)
graph.add_node('summary', summary)


graph.add_edge(START, "calc_sr")
graph.add_edge(START, "calc_bpb")
graph.add_edge('calc_sr', 'summary')
graph.add_edge('calc_bpb', 'summary')
graph.add_edge('summary', END)

workflow = graph.compile()           
```

```python
result = workflow.invoke({'run': 20, 'four':2, 'six':1, 'ball':6})
print(result)
```
